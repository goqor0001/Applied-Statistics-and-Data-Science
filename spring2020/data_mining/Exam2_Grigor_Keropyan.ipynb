### Prepare data for the algorithms (10 points) 

Hint: The input for algorithms should be 'list', the corresponding values should be 'string' and the data should not contain nan values

```python
def f(d):
    l = d.dropna()
    l = l.tolist()
    return l

prep = data.apply(f, axis=1) 
prep = prep.tolist()
```

```python
prep
```

### Print the frequent itemsets if min_Support=0.5% (10 points)

```python
te = TransactionEncoder()
te_ary = te.fit(prep).transform(prep)
df = pd.DataFrame(te_ary, columns=te.columns_)
frequent_itemsets = apriori(df, min_support=0.005, use_colnames=True)

frequent_itemsets
```

### Print the rules derived from the frequent itemsets only if the min_Confidence=25%  (5 points)

```python
rules = association_rules(frequent_itemsets, min_threshold=0.25)
rules
```

### Print the rules that satisfy the following criteria: (10 points)

at least 2 antecedents,
the min_confidence = 30%
and lift score >= 3 

```python
f = rules.antecedents.apply(lambda x: len(x)) >= 2
s = rules.confidence>=0.3
t = rules.lift>=3

rules[f & s & t]
```
